```python
import pandas as pd
import re

# 读取CSV文件
dongcheng_df = pd.read_csv('dongcheng.csv', encoding='GB18030')
haidian_df = pd.read_csv('haidian.csv', encoding='GB18030')
huairou_df = pd.read_csv('huairou.csv', encoding='GB18030')
tongzhou_df = pd.read_csv('tongzhou.csv', encoding='GB18030')

```

```python
# 获取属性名（第一行信息）
columns = dongcheng_df.columns

# 打印属性名
print(columns)
```

```python
# 显示每个数据集的前五行数据
print("Dongcheng 数据集前五行:")
print(dongcheng_df.head())

print("\nHaidian 数据集前五行:")
print(haidian_df.head())

print("\nHuairou 数据集前五行:")
print(huairou_df.head())

print("\nTongzhou 数据集前五行:")
print(tongzhou_df.head())
```

```python
# 保留指定列
selected_columns = ['总价格', '单价', '地段', '房屋户型', '所在楼层', '建筑面积', '户型结构', '装修情况']
dongcheng_df = dongcheng_df[selected_columns]
haidian_df = haidian_df[selected_columns]
huairou_df = huairou_df[selected_columns]
tongzhou_df = tongzhou_df[selected_columns]
```

```python
# 输出处理后的数据的前3行
print("Dongcheng 数据集处理后的前3行:")
print(dongcheng_df.head(3))

print("\nHaidian 数据集处理后的前3行:")
print(haidian_df.head(3))

print("\nHuairou 数据集处理后的前3行:")
print(huairou_df.head(3))

print("\nTongzhou 数据集处理后的前3行:")
print(tongzhou_df.head(3))
```

```python
# 查看每个数据集各属性列的大体情况
print("Dongcheng 数据集各属性列的大体情况:")
print(dongcheng_df.describe())

print("\nHaidian 数据集各属性列的大体情况:")
print(haidian_df.describe())

print("\nHuairou 数据集各属性列的大体情况:")
print(huairou_df.describe())

print("\nTongzhou 数据集各属性列的大体情况:")
print(tongzhou_df.describe())
```

```python
# 去除建筑面积列中的"㎡"字符并转换为float类型
# 将所在楼层()里面的内容删去

# 去除建筑面积列中的"㎡"字符并转换为float类型
dongcheng_df['建筑面积'] = dongcheng_df['建筑面积'].str.rstrip('㎡').astype(float)
haidian_df['建筑面积'] = haidian_df['建筑面积'].str.rstrip('㎡').astype(float)
huairou_df['建筑面积'] = huairou_df['建筑面积'].str.rstrip('㎡').astype(float)
tongzhou_df['建筑面积'] = tongzhou_df['建筑面积'].str.rstrip('㎡').astype(float)

# 使用正则表达式删除括号及其内部的内容
dongcheng_df['所在楼层'] = dongcheng_df['所在楼层'].str.replace(r'\(.*\)', '', regex=True)
haidian_df['所在楼层'] = haidian_df['所在楼层'].str.replace(r'\(.*\)', '', regex=True)
huairou_df['所在楼层'] = huairou_df['所在楼层'].str.replace(r'\(.*\)', '', regex=True)
tongzhou_df['所在楼层'] = tongzhou_df['所在楼层'].str.replace(r'\(.*\)', '', regex=True)
```

```python
# 1. 处理重复值
# 删除重复行，要求所有列都相同才删除
dongcheng_df = dongcheng_df.drop_duplicates(subset=dongcheng_df.columns.tolist(), keep='first')
haidian_df = haidian_df.drop_duplicates(subset=haidian_df.columns.tolist(), keep='first')
huairou_df = huairou_df.drop_duplicates(subset=huairou_df.columns.tolist(), keep='first')
tongzhou_df = tongzhou_df.drop_duplicates(subset=tongzhou_df.columns.tolist(), keep='first')

```

```python

# 2. 处理缺失值
# 删除缺失值
dongcheng_df = dongcheng_df.dropna()
haidian_df = haidian_df.dropna()
huairou_df = huairou_df.dropna()
tongzhou_df = tongzhou_df.dropna()


```

```python
# 查看每个数据集字符类型列的大体情况
print("Dongcheng 数据集各字符类型列的大体情况:")
for column in dongcheng_df.columns:
    if dongcheng_df[column].dtype == 'object':
        print(f"列名：{column}")
        print(dongcheng_df[column].value_counts())
        print("\n唯一值：")
        print(dongcheng_df[column].unique())
        print("\n")

print("\nHaidian 数据集各字符类型列的大体情况:")
for column in haidian_df.columns:
    if haidian_df[column].dtype == 'object':
        print(f"列名：{column}")
        print(haidian_df[column].value_counts())
        print("\n唯一值：")
        print(haidian_df[column].unique())
        print("\n")


# 查看 Huairou 数据集各字符类型列的大体情况
print("Huairou 数据集各字符类型列的大体情况:")
for column in huairou_df.columns:
    if huairou_df[column].dtype == 'object':
        print(f"列名：{column}")
        print(huairou_df[column].value_counts())
        print("\n唯一值：")
        print(huairou_df[column].unique())
        print("\n")
# 查看 Tongzhou 数据集各字符类型列的大体情况
print("Tongzhou 数据集各字符类型列的大体情况:")
for column in tongzhou_df.columns:
    if tongzhou_df[column].dtype == 'object':
        print(f"列名：{column}")
        print(tongzhou_df[column].value_counts())
        print("\n唯一值：")
        print(tongzhou_df[column].unique())
        print("\n")
```

```python
# 保存处理后的数据集到新文件，并指定编码方式
dongcheng_df.to_csv('dongcheng_after.csv', index=False, encoding='GB18030')
haidian_df.to_csv('haidian_after.csv', index=False, encoding='GB18030')
huairou_df.to_csv('huairou_after.csv', index=False, encoding='GB18030')
tongzhou_df.to_csv('tongzhou_after.csv', index=False, encoding='GB18030')

```
